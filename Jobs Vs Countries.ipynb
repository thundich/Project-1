```python
#Dependencies:
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Load Data:
country_df = pd.read_csv("../Project-1/salaries.csv")
country_df
```

```python
#Data info
country_df.info()
```

```python
country = country_df["company_location"].unique()
country
```

```python
Countries = len(country)
Countries
```

```python
Job = country_df["job_title"].unique()
print(Job)
```

```python
Jobs = len(Job)
Jobs
```

```python

country_df.groupby('job_title').company_location.count().sort_values(ascending=False)[:10].plot.bar()

plt.title("Job Title Vs Number of Jobs")
plt.xlabel("Job Title")
plt.ylabel("Number of Jobs")

plt.savefig('Title Vs Jobs.png', transparent=True)
plt.show()
```

```python
y_axis = country_df["job_title"]
x_axis = country_df["company_location"]
plt.figure(figsize = (15,12))

plt.title("Company Location Vs Job Titles")
plt.bar(x_axis, y_axis,linewidth=2, color = "c")
plt.xticks(x_axis, rotation = "vertical")
plt.xlabel("Company Location")
plt.ylabel("Job Title")

plt.savefig("Location Vs Title.png", transparent=True)
plt.show()
```

```python
country_df.groupby('company_location').job_title.count().sort_values(ascending=False)[:10].plot.bar()

plt.title("Company Location Vs Number of Jobs")
plt.xlabel("Company Location")
plt.ylabel("Number of Jobs")
plt.show()
plt.savefig('Company Location Vs Jobs.png', transparent=True)
```

```python
remote = country_df["remote_ratio"].unique()
remote
```

100 = Fully remote work (more that 80%

```python
remote = country_df[country_df['remote_ratio'] == '100']
job_type = country_df.groupby('employment_type').remote_ratio.count()
job_type
```

```python
remote = [18,11,8086,13]
types = ['CT', 'FL', 'FT', 'PT']
colors = ["red", "yellow", "blue","green"]
explode = (0.1,0,0,0)

plt.pie(remote, explode=explode, labels=types, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=90)
plt.axis("equal")
plt.savefig('Remote vs Employment Type.png', transparent=True)
plt.show()
```
