```python
# Dependencies
import json
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as st
from scipy.stats import linregress

# Load JSON
filepath = Path("../Resources/salaries.json")
with open(filepath) as jsonfile:
    salaries_json = json.load(jsonfile)
    
salaries_json
```

```python
# Create a Data Frame and Explore Data  
df = pd.DataFrame.from_records(salaries_json)
df.info()
df.head()

# # Modofy Salary to float
# df["salary_in_usd"] = [float(salary) for salary in  df["salary_in_usd"]]
# df.info()
```

```python
#Check for Null Values
df.isnull().sum()
```

```python
#Average Earning for Position Title and 
bytitles =  df.groupby("job_title")
sal_range = bytitles.mean("salary_in_usd")["salary_in_usd"].reset_index()
sal_range
```

```python
# Plot Bar Chart Salary Vs Comp Size
# sal_range.plot(kind = 'bar', xlabel = "job_title", ylabel = "Salary", figsize = (20,4))

y_axis = sal_range["salary_in_usd"]
x_axis = sal_range["job_title"]
tick_locations = [value for value in x_axis]
plt.figure(figsize = (20,4))

plt.bar(x_axis, y_axis)
plt.xticks(tick_locations, x_axis, rotation = "vertical")
plt.xlabel("Job Title")
plt.ylabel("Salary in USD")
plt.show()
plt.tight_layout()
```

```python
#salary BY Company Size
by_cmop_size = df.groupby("company_size")
# ['salary_in_usd'].reset_index()
comp_size_df = by_cmop_size.mean("salary_in_usd")["salary_in_usd"].reset_index()
comp_size_df.sort_values("salary_in_usd" , ascending  = False)
```

```python
comp_size_df.plot(kind = 'bar', xticks = comp_size_df['company_size'] , xlabel = "company_size", ylabel = "salary_in_usd", figsize = (8,6))
```
